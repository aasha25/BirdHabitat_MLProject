## TASK 2(a): Model Accuracy Graphs using MinMaxScaler()

```python
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score

#setting the dataframe
df = pd.DataFrame({
    'wingspan': [20, 25, 15, 18, 200, 180, 10, 45, 30, 40, 12, 150, 180, 150, 40],
    'lifespan': [5, 7, 3, 6, 20, 15, 3, 15, 10, 15, 8, 25, 20, 25, 8],
    'habitat': ['Forest', 'Forest', 'Other', 'Forest', 'Mountain', 'Mountain', 'Other', 'Other', 'Forest', 'Other', 'Other', 'Other', 'Other', 'Other', 'Other']})

dfX = df[['wingspan', 'lifespan']]
sy = df['habitat']

#splitting into traininsg and testing data
dfX_train, dfX_test, sy_train, sy_test = train_test_split(dfX, sy, test_size=0.2, random_state=42)

# Applying MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(dfX_train)
X_test = scaler.transform(dfX_test)

#choosing a range of k values
k_values = [1, 2, 3, 4, 5, 6, 7, 8, 9]

# Lists to store accuracy values
train_accuracies = []
test_accuracies = []

# for loop to evaluate knn with different k values
for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, sy_train)

    # Predictions on the training set
    ytrain_prediction = knn.predict(X_train)
    train_accuracy = accuracy_score(sy_train, ytrain_prediction)
    train_accuracies.append(train_accuracy)

    # Predictions on the test set
    ytest_prediction = knn.predict(X_test)
    test_accuracy = accuracy_score(sy_test, ytest_prediction)
    test_accuracies.append(test_accuracy)

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(k_values, train_accuracies, label='Train Accuracy', marker='^', color = 'c')
plt.plot(k_values, test_accuracies, alpha=0.5, label='Test Accuracy', marker='o', color = 'purple')
plt.title('Model Accuracy vs. k Values Using MinMaxScaler()')
plt.xlabel('k Values')
plt.ylabel('Model Accuracy')
plt.legend()
plt.grid(True)
plt.show()
```

## TASK 3(b): Creating and explaining the Confusion Matrix of the model

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.metrics import accuracy_score, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

#setting the dataframe
df = pd.DataFrame({
    'wingspan': [20, 25, 15, 18, 200, 180, 10, 45, 30, 40, 12, 150, 180, 150, 40],
    'lifespan': [5, 7, 3, 6, 20, 15, 3, 15, 10, 15, 8, 25, 20, 25, 8],
    'habitat': ['Forest', 'Forest', 'Other', 'Forest', 'Mountain', 'Mountain', 'Other', 'Other', 'Forest', 'Other', 'Other', 'Other', 'Other', 'Other', 'Other']})


dfX = df[['wingspan', 'lifespan']]
sy = df['habitat']


#splitting into traininsg and testing data
dfX_train, dfX_test, sy_train, sy_test = train_test_split(dfX, sy, test_size=0.2, random_state=42)

# Choosing the best k and scaler
best_k = 1
best_scaler = StandardScaler()

X_train = best_scaler.fit_transform(dfX_train)
X_test = best_scaler.transform(dfX_test)

knn = KNeighborsClassifier(n_neighbors=best_k)
knn.fit(X_train, sy_train)

# Predictions on the training and test dataset
ytrain_prediction = knn.predict(X_train)
ytest_prediction = knn.predict(X_test)

# Calculating accuracy
train_accuracy = accuracy_score(sy_train, ytrain_prediction)
print(f"Training Accuracy: {train_accuracy:.4f}")

test_accuracy = accuracy_score(sy_test, ytest_prediction)
print(f"Test Accuracy: {test_accuracy:.4f}")

# Calculating confusion matrix
confused_matrix = confusion_matrix(sy_test, ytest_prediction)

# Plotting the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(confused_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=knn.classes_, yticklabels=knn.classes_)
plt.title('Confusion Matrix for Test Dataset')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

#### Explaining the Confusion Matrix:
The confusion matrix code evaluates a knn model on a test dataset. The confusion matrix illustrates the model's classification performance, depicting true positives, true negatives, false positives, and false negatives. The heatmap visually represents the counts, aiding in the assessment of class-specific prediction accuracy. The code achieves this using a standard scaler, selecting optimal parameters, and visualizing the confusion matrix to assess the knn model's effectiveness in classifying habitat types.

#### Interpretation the Confusion Matrix
The confusion matrix provides insights into the performance of a classification model, particularly in terms of how well it predicts each class. In the context of the model developed:

True Positives (TP): The number of instances correctly predicted as belonging to the 'Forest' or 'Mountain' habitat is 1.
True Negatives (TN): The number of instances correctly predicted as not belonging to the 'Forest' or 'Mountain' habitat is 2.
False Positives (FP): The number of instances incorrectly predicted as belonging to the 'Forest' or 'Mountain' habitat when they actually belong to another habitat is 0.
False Negatives (FN): The number of instances incorrectly predicted as not belonging to the 'Forest' or 'Mountain' habitat when they actually do is 0.
